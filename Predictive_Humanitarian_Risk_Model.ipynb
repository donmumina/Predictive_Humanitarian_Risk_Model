```python
import pandas as pd

food = pd.read_csv("wfp_food_prices_sdn.csv")
food.head()
```

```python
# Standardize column names
food.columns = food.columns.str.lower().str.strip()

# Convert date
food['date'] = pd.to_datetime(food['date'])

# Keep relevant columns only
food = food[
    ['date', 'admin1', 'admin2', 'market',
     'commodity', 'price', 'usdprice']
]

# Drop rows with missing prices
food = food.dropna(subset=['usdprice'])
```
