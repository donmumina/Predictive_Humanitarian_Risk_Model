## 4Ô∏è‚É£ MODELING  

### ‚úÖ Modeling Strategy (Unsupervised + Index-Based Modeling)  
This project does **not** apply supervised machine learning models such as **Random Forest**, **K-Nearest Neighbors (KNN)**, or **Logistic Regression** due to the absence of **ground-truth outcome labels** (e.g., IPC phases, nutrition survey indicators, household vulnerability scores).

Instead, the project applies **unsupervised modeling** to detect early warning signals and construct an interpretable humanitarian risk score.



### üìå Model 1: Unsupervised Anomaly Detection (Early Warning Signals)  
Unsupervised anomaly detection is used to flag abnormal shifts in key indicators such as:

- rapid food price inflation  
- volatility spikes  
- conflict intensity surges  
- unusual displacement growth patterns  

These anomalies form the primary **early warning alerts** to support **anticipatory action**.



### üìå Model 2: Composite Risk Scoring (Humanitarian Risk Index ‚Äî HRI)  
A **Humanitarian Risk Index (HRI)** is constructed by integrating normalized indicators from:

- market stress *(inflation & volatility)*  
- conflict intensity *(events & fatalities)*  
- displacement pressure *(macro trends)*  

The HRI produces:

‚úÖ monthly risk scores  
‚úÖ risk categories *(LOW / MODERATE / HIGH)*  
‚úÖ geographic prioritization outputs for planning and response  



### üß† Why Unsupervised Methods Were Chosen  
Unsupervised modeling is appropriate because it:

- works without labeled outcomes  
- supports early warning through abnormal pattern detection  
- produces explainable signals suitable for humanitarian decision-making  
- aligns with real-world early warning approaches used by humanitarian agencies

## Creating `refugees_origin_level`

This code creates a simplified refugees dataset for analysis.

- It selects only 5 important columns:  
  `pop_origin_name`, `date`, `month`, `year`, `individuals`

- It converts columns into the correct formats:
  - `date` ‚Üí datetime  
  - `month`, `year`, `individuals` ‚Üí numeric

- It prints the dataset shape and shows the first 5 rows to confirm the data loaded correctly.


````python
refugees_origin_level = refugees_df[["pop_origin_name", "date", "month", "year", "individuals"]].copy()

# Convert types
refugees_origin_level["date"] = pd.to_datetime(refugees_origin_level["date"], errors="coerce")
refugees_origin_level["month"] = pd.to_numeric(refugees_origin_level["month"], errors="coerce")
refugees_origin_level["year"] = pd.to_numeric(refugees_origin_level["year"], errors="coerce")
refugees_origin_level["individuals"] = pd.to_numeric(refugees_origin_level["individuals"], errors="coerce")

print(" refugees_origin_level created:", refugees_origin_level.shape)
display(refugees_origin_level.head())
````

## Creating `refugees_monthly_pressure`

This code creates a monthly displacement pressure indicator.

- It groups the origin-level dataset by `date`
- It sums the `individuals` column to get the total displaced individuals for that reporting period
- It renames the summed column to `total_displaced_individuals`
- The result is a dataset with **one row per date**, showing total displacement pressure



````python
refugees_monthly_pressure = (
    refugees_origin_level.groupby("date")["individuals"]
    .sum()
    .reset_index()
    .rename(columns={"individuals": "total_displaced_individuals"})
)

print(" refugees_monthly_pressure created:", refugees_monthly_pressure.shape)
display(refugees_monthly_pressure)
````

## Converting Food Prices Date Column and Creating Monthly Key

The food prices dataset includes a `date` column stored as text.  
To support time-series analysis, we:

1. Convert `date` into datetime format
2. Create a new `month` column that groups records by year-month

This monthly key will allow:
- aggregation of prices per market and commodity
- inflation and volatility calculations
- alignment with conflict and displacement indicators


````python
# Convert date column to datetime
food_df["date"] = pd.to_datetime(food_df["date"], errors="coerce")

# Create month key for aggregation
food_df["month"] = food_df["date"].dt.to_period("M").dt.to_timestamp()

print(" Food price date converted and month key created!")
food_df[["date", "month"]].head()
````
